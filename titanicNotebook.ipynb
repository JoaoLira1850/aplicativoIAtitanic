```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import joblib
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
df.head()
```

```python
df.info()
```

```python
df['Age'].median()
```

```python
df['Age'].fillna ( df["Age"].median(), inplace= True)
```

```python
df.info()
```

```python
#Criaçao de variaveis Dummies da coluna Sex
```

```python
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
df.head()
```

```python
#criação de variaveis dummis da coluna Pclas(usando get_dummies)
```

```python
df = pd.get_dummies(df,columns = ['Pclass','Embarked'],drop_first=True, dtype=int)
df.head()
```

```python
#Distretização  de variaveis continuas (age e Fare)

#Age
bins = [0,12,18,50,100]
labels = ['crianca','adolescente','adulto','idoso']
df['faixaEtaria'] = pd.cut(df['Age'], bins = bins, labels = labels)

#Fare
bins = [0,7.91,14,454.31,512]
labels = ['Low','Mid-low','Mid-high','High']

df['faixaTarifa'] = pd.cut(df['Fare'], bins = bins, labels = labels)

```

```python
#Ciriação de variaveis dummis da colunas para FaixaEtaria e faixaTarifa
df = pd.get_dummies(df, columns=['faixaEtaria', 'faixaTarifa'], drop_first=True, dtype=int)
```

```python
#Criar as entradas X e Y
df.columns
```

```python
X = df[['Sex', 'Age', 'SibSp', 'Parch','Fare', 'Pclass_2', 'Pclass_3', 'Embarked_Q','Embarked_S']]
Y = df['Survived']
```

```python
# Separar os conjuntos em treinamento e teste
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.1,random_state = 42)

```

```python
#modelo de regreção logistica para dataset do titanic
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
#fazer previsão
Y_pred = model.predict(X_test)
```

```python
#Metricas de performance do modelo de regressão Logistica para o dataset do titanic
relatorio = classification_report(Y_test, Y_pred)
print(f'relatorio de classificação:\n{relatorio}')
```
